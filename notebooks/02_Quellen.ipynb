# Erstelle Liste von Quellen und Schrifte

Als Input dient [diese Seite]() mit einer Liste von Quellen und Schriften, die im WWb referenziert werden.

```python
import pandas as pd
import regex as re
```

```python
# Open the raw Data
with open('../data/raw/Quellen.html', 'r', encoding='latin-1') as f:
    html_content = f.read()
```

```python
# Replace a few things
html_content = re.sub(r'', '–', html_content) # The first one is NOT an empty string!
html_content = re.sub(r'<sup class="preface-superscript">1</sup>', '¹', html_content)
html_content = re.sub(r'<sup class="preface-superscript">2</sup>', '²', html_content)
html_content = re.sub(r'<sup class="preface-superscript">3</sup>', '³', html_content)
html_content = re.sub(r'<sup class="preface-superscript">4</sup>', '⁴', html_content)
html_content = re.sub(r'<sup class="preface-superscript">5</sup>', '⁵', html_content)
html_content = re.sub(r'<sup class="preface-superscript">6</sup>', '⁶', html_content)
html_content = re.sub(r'<sup class="preface-superscript">7</sup>', '⁷', html_content)
html_content = re.sub(r'<sup class="preface-superscript">8</sup>', '⁸', html_content)
html_content = re.sub(r'<sup class="preface-superscript">9</sup>', '⁹', html_content)
html_content = re.sub(r'<sup class="preface-superscript">10</sup>', '¹⁰', html_content)
html_content = re.sub(r'<sup class="preface-superscript">11</sup>', '¹¹', html_content)
html_content = re.sub(r'<sup class="preface-superscript">12</sup>', '¹²', html_content)
html_content = re.sub(r'<sup class="preface-superscript">\*</sup>', '*', html_content)
html_content = re.sub(r'<sup class="preface-superscript">1\+2</sup>', '¹⁺²', html_content)
html_content = re.sub(r'<br/>', '', html_content)
html_content = re.sub(r'\n(?!<dt)', r' ', html_content) # The order of these last two matters!
html_content = re.sub(r'<span class=""?preface-italics""?>(.+?)</span>', r'\1', html_content)
```

```python
# Split up lines
html_lines = html_content.split('\n')
for line in html_lines:
    print(line)
```

```python
# Lists for holding the data that will later make up the dataframe
quelle_abbreviations = []
quelle_descriptions = []

for line in html_lines:
    new_quelle = re.findall(r'<dt class="preface-definitionterm">(.+?)</dt><dd class="preface-definition">(.+?)</dd>', line)
    if len(new_quelle):
        quelle_abbreviations.append(new_quelle[0][0])
        quelle_descriptions.append(new_quelle[0][1])

# Combine the lists created above into dataframes
quellen = pd.DataFrame({'Abkuerzung': quelle_abbreviations, 'Beschreibung': quelle_descriptions})


# Save the dataframes as CSVs
#quellen.to_csv('../data/processed/Quellen.csv', index=False, sep='\t')
```
