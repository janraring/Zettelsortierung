### Adding Unspecified Places

```python
kreise_df = pd.read_csv('../data/processed/Kreise.csv', delimiter='\t')
kreise_df
```

```python
def unbestimmt(name: str) -> str:
    if name[0] == "K":
        return f"Unbestimmter Ort im {name}"
    if name[:15] == "die krfr. Stadt":
        return f"Unbestimmter Ort in der {name[4:]}"
    if name[:15] == "die krfr. Städt":
        return f"Unbestimmter Ort in den krfr. Städten {name[17:]}"
    if name[0] == "f":
        return f"Unbestimmter Ort im früheren {name[8:]}"
    return f"Unbestimmter Ort in {name}"

kreise_df["Name"] = kreise_df["Name"].map(lambda n: unbestimmt(n))
kreise_df
```

```python
abk_name = list(kreise_df.itertuples(index=False, name=None))
unbest_orte = [(abk, "Xy", name) for abk, name in abk_name]
unbest_orte
```

```python
db.add_orte(unbest_orte)
```
