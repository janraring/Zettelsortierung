### Adding Zettel

```python
odd_paths = [path for path in paths_list if len(re.findall('1#', path))]
print(len(odd_paths))
print(odd_paths[0])
```

```python
zettel_list = [Zettel(path) for path in odd_paths]
print(zettel_list[0])
```

```python
db.add_zettel_list(zettel_list)
```
