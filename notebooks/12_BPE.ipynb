# Byte-Pair Endocing on the OCR-results

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.normalizers import Lowercase, NFD, StripAccents, Sequence
import regex as re
from collections import Counter

from zettelsortierung.DataModel import DataBase

db = DataBase()
ocr_results = db.get_ocr_line_by_line()
```

```python
by_lines = [re.sub('Æ', '\n', res[0]) for res in ocr_results]
print(by_lines[1])
concat = '\n'.join(by_lines)
print(concat[:1000])
```

```python
with open('input/concat_ocr.txt', 'w') as f:
    f.write(concat)
```

```python
# 1. Initialize a tokenizer with a BPE model
tokenizer = Tokenizer(BPE(unk_token="[UNK]"))

# 2. Optional: Add normalization (lowercasing, accent stripping, etc.)
#tokenizer.normalizer = Sequence([NFD(), Lowercase(), StripAccents()])

# 3. Pre-tokenize on whitespace
tokenizer.pre_tokenizer = Whitespace()

# 4. Setup BPE trainer
trainer = BpeTrainer(
    vocab_size=1000,
    min_frequency=2,
    special_tokens=["[UNK]"],
)

# 5. Train on your text file(s)
files = ["input/concat_ocr.txt"]  # Replace with your file path
tokenizer.train(files, trainer)
#tokenizer.train_from_iterator(concat, trainer=trainer)

# 6. Save tokenizer
tokenizer.save("output/bpe_tokenizer.json")

print("Training complete. Tokenizer saved to bpe_tokenizer.json")
```

```python
from transformers import PreTrainedTokenizerFast

tokenizer = Tokenizer.from_file('output/bpe_tokenizer.json')


with open('input/concat_ocr.txt', 'r') as f:
    concat_ocr = f.read()
encoded = tokenizer.encode(concat_ocr)
```

```python
counts = dict(Counter(encoded.ids).most_common())
```

```python
print(counts)
for key, value in counts.items():
    print(f'{tokenizer.decode([key])} ({value})')
```
